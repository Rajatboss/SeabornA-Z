```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
print(df.head())
```

```python
#to view all the data available in seaborn library
print(sns.get_dataset_names())
```

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5): 
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
plt.show()
```

```python
# To change the same plot to Seaborn defaults, use the set() function −
sns.set()
sinplot()
plt.show()
```

```python
# Basically, Seaborn splits the Matplotlib parameters into two groups−

# Plot styles
# Plot scale

# Seaborn Figure Styles
# The interface for manipulating the styles is set_style().

sns.set_style("whitegrid")
sinplot()
# plt.show()
```

```python
# Removing Axis Spine
# In the white and ticks themes, we can remove the top and right axis spines using the despine() function
  
sns.set_style("white")
sinplot()
sns.despine()
```

```python
print(sns.axes_style())
```

```python
sns.set_style("darkgrid", {'axes.axisbelow': False})
sinplot()
sns.despine()
```

```python
# Scaling Plot Elements
# sns.set_context('paper')
sns.set_context('poster')
sns.set_style("darkgrid",{'axes.axisbelow':False})
sinplot()
```

```python
#Seaborn - Color Palette
sns.color_palette(palette=None, n_colors=None, desat=None)
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
# Sequential Color Palettes
# sequential plots are suitable to express the distribution of data ranging from relative lower values to higher values within a range
sns.palplot(sns.color_palette("Purples"))
plt.show()
```

```python
# Diverging Color Palette:- Diverging palettes use two diffenet colors.
# Asuming plotting the data ranging from -1 to 1. The values from -1 to 0 takes one color and 0 to +1 take another color
sns.palplot(sns.color_palette("BrBG",11))
```
