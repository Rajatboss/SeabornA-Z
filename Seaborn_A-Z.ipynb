```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5): 
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
plt.show()
```

### There are 3 types of palettes : sequential, diverging, and qualitative.

<hr>
<br>
<div style="color:green"> Sequential palettes are suited to ordered data that progress from low to high (gradient). The palettes names are : Blues, BuGn, BuPu, GnBu, Greens, Greys, Oranges, OrRd, PuBu, PuBuGn, PuRd, Purples, RdPu, Reds, YlGn, YlGnBu YlOrBr, YlOrRd.</div>
<br>
<div style="color:green"> Diverging palettes put equal emphasis on mid-range critical values and extremes at both ends of the data range. The diverging palettes are : BrBG, PiYG, PRGn, PuOr, RdBu, RdGy, RdYlBu, RdYlGn, Spectral</div>
<br>
<div style="color:green"> Qualitative palettes are best suited to representing nominal or categorical data. They not imply magnitude differences between groups. The palettes names are : Accent, Dark2, Paired, Pastel1, Pastel2, Set1, Set2, Set3</div>

```python
# Qualitavtive Color Palettes
# Qualitative or categorical palettes are best suitable to plot the categorical data.
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
# Sequential Color Palettes
# sequential plots are suitable to express the distribution of data ranging from relative lower values to higher values within a range
sns.palplot(sns.color_palette("Greens"))
plt.show()
```

```python
# Diverging Color Palette:- Diverging palettes use two diffenet colors.
# Diverging palettes use two different colors. Each color represents variation in the value ranging from a com mon point in either direction.
sns.palplot(sns.color_palette("PiYG",11))
```

```python
# Plotting Univariate Distribution
# Distribution of data s the most important thing we need to understand while analysing the data 
# distplot() provides the most convenient way to take a quick look at univariate distribution
# this function will plot a histogram that fits the kernal density estimation of data
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'],kde=False)
plt.show()
```

```python
#Linear Relationship
# Functions to Draw Linear Regression Models
#                 regplot                                            lmplot
# accepts the x and y variables in a variety of      has data as a required parameter and the x
# form ats including sim ple num py a rrays,         and y variables m ust be specified as strings.
# pandas Series objects, or as references to         This data form at is called “long -form ” data
# variables in a pandas DataFrame
```

```python
df = sns.load_dataset('tips')
sns.regplot(x="total_bill", y="tip", data=df)
sns.lmplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# W e can also fit a linear regression when one of the variables takes discrete values
df = sns.load_dataset('tips')
sns.lmplot(x="size", y="tip", data=df)
plt.show()
```

```python
# Fitting Different Kinds of Models
# The sim ple linear regression m odel used above is very sim ple to fit, but in m ost of the
# cases, the data is no n-linear and the above m ethods cannot generalize the regression
# line.
df = sns.load_dataset('anscombe')
sns.lmplot(x="x", y="y", data=df.query("dataset == 'I'"))
plt.show()
```

```python
sns.lmplot(x='x', y='y', data=df.query("dataset == 'II'"))
plt.show()
```

```python
# The above plot is not a good linear regression plot so we can
# choose to plot non-linear regression for such data
sns.lmplot(x='x',y='y',data=df.query("dataset == 'II'"), order=2)
plt.show()
```

```python
# Facet Grid
# A useful approach to explore medium-dimensional data, is by drawing multiple instances of the same plot on different sunsets od your dataset
# This technique is commonly called as 'lattice', or 'trells' plotting and it is related to the idea of 'small multipes'
# FacetGrid object takes a datafram e as input and the nam es of the variables that will
# form the row, colum n, or hue dim ensions of the grid.
# The variables should be categorical and the data at each level of the variable will be used
# for a facet along that axis.
```

```python
df = sns.load_dataset('tips')
g = sns.FacetGrid(df, col='time')
plt.show()
```

```python
# I n the above example, we have just initialized the facetgrid object which doesn’t draw
# anything on them .
# The m ain approach for visualizing data on this grid is with the FacetGrid.map()
# m ethod.
g = sns.FacetGrid(df, col='time')
g.map(plt.hist, 'tip')
plt.show()
```

```python
# To m ake a relational plot, pass the m ultiple variable nam es.
g = sns.FacetGrid(df, col='sex', hue='smoker')
g.map(plt.scatter, 'total_bill','tip')
plt.show()
```

```python
# PairGrid allows us to draw a grid of subplots using the sam e plot ty pe to visualize data.

# Unlike FacetGrid, it uses different pair of variable for each subplot. I t form s a m atrix of
# sub-plots. I t is also som etim es called as “scatterplot m atrix”.

# The usage of pairgrid is sim ilar to facetgrid. First initialise the grid and then pass the
# plotting function.
```

```python
df = sns.load_dataset('iris')
g = sns.PairGrid(df)
g.map(plt.scatter)
plt.show()
```

```python
# I t is also possible to plot a different function on the diagonal to show the univariate
# distribution of the variable in each colum n.
g = sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
plt.show()
```

```python
# W e can custom ize the color of these plots using another categorical variable. For
# exam ple, the iris dataset has four m easurements for each of three different species of
# iris flowers so you can see how they differ.
g = sns.PairGrid(df, hue='species')
g.map_diag(plt.hist, color='green')
g.map_offdiag(plt.scatter)
g.add_legend()
plt.show()
```

```python
# We can use a different function in the u pper and lower triangles to see different aspects of the relationship.
g = sns.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.map_diag(sns.kdeplot, lw=3, legend=False)
plt.show()
```
